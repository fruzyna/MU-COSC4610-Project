```python
def removeCounty(county):
    return county.replace(' County', '')
```

```python
poverty = pd.read_csv('data/est16-ia.csv')
poverty['County'] = poverty['County'].apply(removeCounty)
poverty = poverty.set_index('County')
income = poverty['Median_Household_Income']
income
```

```python
%matplotlib inline
income.plot(kind='bar')
```
